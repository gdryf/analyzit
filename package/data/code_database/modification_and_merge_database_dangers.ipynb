```python
import pandas as pd

def modification_document_csv(name_file_entry, name_file_exit):
    # To read the csv file
    df = pd.read_csv(name_file_entry)

    # List of the columns taht need to be conserved int he new document
    columns_conserved = ["cmpdname", "cmpdsynonym", "iupacname"]

    # Suppresion of all the other columns of the initial file
    columns_suppression = [colonne for colonne in df.columns if colonne not in columns_conserved]
    df.drop(columns=columns_suppression, inplace=True)

    # Save of the new csv document 
    df.to_csv(name_file_exit, index=False)

def verify_columns(name_file_exit): # test to check that the we have all the informations in the new csv file
    # To read the new csv file
    df = pd.read_csv(name_file_exit)

    # List of the columns that are supposed to be in the document 
    columns_normal = ["cmpdname", "cmpdsynonym", "iupacname"]

    # Verification that the columns are correct
    if list(df.columns) == columns_normal:
        return("The columns of this document are correct.")
    else:
        return("The columns of this document are not correct.")


    
# Example of the use of the function 
name_file_entry = "PubChem_compound_text_paraben.csv"
name_file_exit = "PubChem_paraben_final.csv"

modification_document_csv(name_file_entry, name_file_exit)
verify_columns(name_file_exit)
```

```python
import pandas as pd

def add_column_danger(name_danger, file_csv):
    
    df = pd.read_csv(file_csv)
    
    df['dangers'] = name_danger
    
    df.to_csv(file_csv, index=False)

# Example of use
name_danger = "Carcinogenic"
file_csv = "PubChem_carcinogenic_final_precised.csv"
add_column_danger(name_danger, file_csv)

def merge_files_csv(files_csv, file_exit):
    # Create a list to stock the DataFrames of each file
    dfs = []

    # Charge each csv file in a Dataframe and add it to the list
    for file_csv in files_csv:
        df = pd.read_csv(file_csv)
        dfs.append(df)

    # Merge the DataFrames
    df_merged = pd.concat(dfs)

    df_merged = df_merged.groupby('cmpdname').agg(lambda x: ', '.join(set(x.dropna())))

    df_merged = df_merged.reset_index()

    df_merged.to_csv(file_exit, index=False)

# Example of use
files_csv = ["PubChem_carcinogenic_final_precised.csv", "PubChem_endocrine_final_precised.csv", "PubChem_paraben_final_precised.csv"]
file_exit = "database_danger_complete.csv"
merge_files_csv(files_csv, file_exit)
```

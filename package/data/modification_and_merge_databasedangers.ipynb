```python
import pandas as pd

def modifier_document_csv(nom_fichier_entree, nom_fichier_sortie):
    # Permet de lire le fichier csv : il doit être dans le meme dossier que le programme pour que la lecture soit possible
    df = pd.read_csv(nom_fichier_entree)

    # Liste des colonnes à conserver
    colonnes_a_conserver = ["cmpdname", "cmpdsynonym", "iupacname"]

    # Suppression de toutes les autres colonnes : celles qui ne sont pas dans la liste 
    colonnes_a_supprimer = [colonne for colonne in df.columns if colonne not in colonnes_a_conserver]
    df.drop(columns=colonnes_a_supprimer, inplace=True)

    # Enregistrement du nouveau fichier modifié comportant uniquement les colonnes spécifiées
    df.to_csv(nom_fichier_sortie, index=False)

def verifier_colonnes(nom_fichier_sortie): # test qui permet que le document sortant contient les bonnes informations (bonnes colonnes)
    # Lecture du fichier CSV modifié
    df = pd.read_csv(nom_fichier_sortie)

    # Liste des colonnes attendues
    colonnes_attendues = ["cmpdname", "cmpdsynonym", "iupacname"]

    # Vérification que les colonnes sont correctes
    if list(df.columns) == colonnes_attendues:
        print("Les colonnes du document sont correctes.")
    else:
        print("Les colonnes du document ne sont pas correctes.")


    
# example : 
#nom_fichier_entree = "PubChem_compound_text_paraben.csv"
#nom_fichier_sortie = "PubChem_paraben_final.csv"
```

```python
import pandas as pd

def ajouter_colonne_danger(nom_danger, fichier_csv):
    # Charger le fichier CSV
    df = pd.read_csv(fichier_csv)
    
    # Ajouter une nouvelle colonne "danger" avec le nom du danger spécifié
    df['dangers'] = nom_danger
    
    # Enregistrer le DataFrame modifié dans le fichier CSV
    df.to_csv(fichier_csv, index=False)

# Exemple d'utilisation
nom_danger = "Carcinogenic"
fichier_csv = "PubChem_carcinogenic_final_precised.csv"
ajouter_colonne_danger(nom_danger, fichier_csv)


def fusionner_fichiers_csv(fichiers_csv, fichier_sortie):
    # Créer une liste pour stocker les DataFrames de chaque fichier
    dfs = []

    # Charger chaque fichier CSV dans un DataFrame et l'ajouter à la liste
    for fichier_csv in fichiers_csv:
        df = pd.read_csv(fichier_csv)
        dfs.append(df)

    # Fusionner les DataFrames en un seul en utilisant la colonne "cmpdname"
    df_merged = pd.concat(dfs)

    # Regrouper les données par la colonne "cmpdname" et fusionner les autres colonnes (si elles sont les mêmes, sans faire de doublons, si elles sont différentes, en séparant les informations par des virgules)
    df_merged = df_merged.groupby('cmpdname').agg(lambda x: ', '.join(set(x.dropna())))

    # Réinitialiser l'index pour obtenir un DataFrame plat
    df_merged = df_merged.reset_index()

    # Enregistrer le DataFrame fusionné dans un nouveau fichier CSV
    df_merged.to_csv(fichier_sortie, index=False)

# Exemple d'utilisation
#fichiers_csv = ["PubChem_carcinogenic_final_precised.csv", "PubChem_endocrine_final_precised.csv", "PubChem_paraben_final_precised.csv"]
#fichier_sortie = "database_danger_complete.csv"
#fusionner_fichiers_csv(fichiers_csv, fichier_sortie)
```
